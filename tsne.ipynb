```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
# import matplotlib.cm
import matplotlib.pyplot as plt

# 
digits = load_digits()
X = digits.data
y = digits.target

# t-SNE
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X)

# 
plt.figure(figsize=(10, 5))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap=plt.cm.get_cmap('jet', 10))
plt.colorbar(ticks=range(10))
plt.clim(-0.5, 9.5)
plt.show()
```

```python
import torch

key_states = torch.load('./states/key_states.pt').cpu()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

h = 0

tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(key_states[0, h])

n = X_tsne.shape[0]
```

```python

# plot :n//3, :2*n//3, :n
plt.figure(figsize=(10, 5))
color = ['r', 'g', 'b']
for i in range(3):
    X_tsne = tsne.fit_transform(key_states[0, h, i*n//3:(i+1)*n//3])
    plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=color[i], label=f'{i*n//3}:{(i+1)*n//3}')


plt.legend()
plt.show()

```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

for h in [0, 15, 31]:

    tsne = TSNE(n_components=2, random_state=0)
    X_tsne = tsne.fit_transform(key_states[0, h])

    n = X_tsne.shape[0]

    # plot :n//3, :2*n//3, :n
    plt.figure(figsize=(10, 5))
    plt.scatter(X_tsne[:n//3, 0], X_tsne[:n//3, 1], c='r', label=':seq_len//3')
    plt.scatter(X_tsne[n//3:2*n//3, 0], X_tsne[n//3:2*n//3, 1], c='g', label=':2*seq_len//3')
    plt.scatter(X_tsne[2*n//3:, 0], X_tsne[2*n//3:, 1], c='b', label=':seq_len')
    plt.legend()
    plt.title(f'head {h}')
    plt.show()
```
